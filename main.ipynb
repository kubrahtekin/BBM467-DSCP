# Feature Engineering

```python
#feature selection
df.drop(["High","Low","Volume","Dividends","Stock Splits"],axis=1,inplace=True)
```

```python
df['Price Change'] = df.apply(lambda row: 0 if row.Close -
                                  row.Open < 0 else 1 ,  axis = 1)
```

```python
df["event"] = ''
i = 0
for ind in df.index:
    if (df['Date'][ind] in  list(lunar["Date"])):
        df['event'][ind] = "lunar"
    elif (df['Date'][ind] in list(solar["Date"])):
        df['event'][ind] = "solar"
    else:
        df['event'][ind] = ""
```
