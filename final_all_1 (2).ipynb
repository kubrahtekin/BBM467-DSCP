## Analyzing<a class="anchor" id="anl"></a>

In this part, we test the data with the help of Pearson's Chi-Square statistical hypothesis after calculating the increase or decrease in bitcoin prices on the dates of astronomical events.

```python
# This function adds two columns to the given data frame; the amount of the BTC price change and increase or decrease.
def final_df(event, btc):
  changes_price = []
  inc_or_dec = []     # 1(increase) and 0(decrease)
  btc.set_index("Date",inplace=True)
  for i in range(len(event)):
    changes_price.append((btc["Close"][event["END DATE"][i]]-btc["Open"][event["START DATE"][i]]))
    if ((btc["Close"][event["END DATE"][i]]-btc["Open"][event["START DATE"][i]])>0):
      inc_or_dec.append(1)
    else:
      inc_or_dec.append(0)
  df1 = pd.DataFrame(list(zip(changes_price,inc_or_dec)), columns =['Changes in Price', 'Inc or Dec'])
  new_df = pd.concat([event["START DATE"],event["END DATE"],df1,event["TYPE"]],axis=1)
  return new_df
```

```python
# calculated and added changes in price, increase or decrease 
ms_df = final_df(meteor_shower_df,price.copy(deep=True))
ms_df
```

```python
e_df = final_df(equinox_df ,price.copy(deep=True))
e_df
```

```python
se_df = final_df(solar_eclipse_df,price.copy(deep=True))
se_df
```

```python
le_df = final_df(lunar_eclipse_df,price.copy(deep=True))
le_df
```

```python
# concatenation of 4 data frames for rest of analyzing
le_df["TYPE"] = 0    # lunar eclipse
se_df["TYPE"] = 1    # solar eclipse
ms_df["TYPE"] = 2    # meteor shower
e_df["TYPE"] = 3     # equinox
last = pd.concat([le_df, se_df, ms_df, e_df])
last.reset_index(drop=True)
last
```

We created a single data frame that holds all the data and calculated the contingency table using the "Inc or Dec" and "TYPE" columns in this data frame.

```python
crosstab = pd.crosstab(last['Inc or Dec'],last['TYPE'], margins=True)
crosstab
```

Pearson's Chi-Square statistical hypothesis is used to test independence between categorical variables are shown as row and column variables in the contingency table. Independence means that knowing the value of the column variable does not change the probabilities of the row variable (and vice versa).

**Hypothesis :**

>H0(Null Hypothesis) : There is no relation between astronomical events and bitcoin price.

>H1(Alternative Hypothesis) : There is a relation between astronomical events and bitcoin price.


We used the chi2_contingency() function of scipy.stats module takes as input, the contingency table in 2d array format. It returns a tuple containing test statistics, the p-value, degrees of freedom and expected table in that order. The expected table is expected frequencies, based on the marginal sums of the table.

```python
from scipy.stats import chi2_contingency
stat, p, dof, expected = chi2_contingency(crosstab.iloc[0:2,0:4])
print("Chi-square value: ",stat)
print("p-value: ",p)
print("Degree of Freedom: ",dof)
print("Expected Table:\n",expected)
```

We used the chi2.ppf() function of scipy.stats module takes as input, the confidence level and degree of freedom in that order. It returns the critical value.

```python
from scipy.stats import chi2
critical_value= chi2.ppf(q = 0.95, # q= (1-.05) 95% confidence*
                      df = 3)      # df= degree of freedom
print("Critical value:",critical_value)
```

There are two methods for verifying the hypothesis:

1- Using p-value:

If the p-value is greater than the significance value(a=0.05), then H0 is accepted. Otherwise, H1 is accepted.

>0.8858854921230519 > 0.05

>p-value > a-value ⇒ H0 is accepted.

2- Using chi-square value:

>chi-square value(X^2)  = ∑ ( (observed-expected)^2 / expected )

If the critical value is greater or equal to the chi-square value, then H0 is accepted. Otherwise, H1 is accepted.

>7.814727903251179 > 0.6457419833853995

>critical value > chi-square value ⇒ H0 is accepted.

## Conclusion<a class="anchor" id="conc"></a>

In both methods, we found the same solution which is “𝐻0 is accepted “. As a result, we can say that there is no relation between astronomical events and bitcoin prices. Therefore, astronomical events such as eclipses, meteor showers, and equinoxes do not affect bitcoin prices.
